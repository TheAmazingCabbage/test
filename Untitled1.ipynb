```python
import requests
import random
import time
import json


totalscanned = []
newscan = []
testedaccounts = 1
f = open("accounts.txt","w")
def getFriends(id):
    try:
        items = requests.get("https://friends.roblox.com/v1/users/"+str(id)+"/friends", timeout=5)
        listOfFriends = []
        if items.ok:
            if items.json().get("data"):
                items = items.json().get("data")
                for i in items:
                    listOfFriends.append(i.get("id"))
            return listOfFriends
        else:
            time.sleep(5)
            getFriends(id)
    except:
        time.sleep(5)
        getFriends(id)

def scan(loop,startid):
    global newscan
    global totalscanned
    global testedaccounts
    newscan = getFriends(startid)
    if type(newscan) is list:
        totalscanned = totalscanned + newscan
        print(len(totalscanned))

    for i in range(loop - 1):
        for i in newscan or range(0):
            testedaccounts += 1
            friends = getFriends(i)
            if type(friends) is list:
                newscan = newscan + friends
                totalscanned = totalscanned + newscan
                totalscanned = list(dict.fromkeys(totalscanned))
                f.write(json.dumps(totalscanned))
                print(len(totalscanned))
scan(3,1083400611)
print("Tested Accounts: "+str(testedaccounts))
print("Scanned Accounts: "+str(len(totalscanned)))
print("Avg Friend Count: "+str(len(totalscanned) / testedaccounts))
```
